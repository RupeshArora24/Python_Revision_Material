```python
import random
```

```python
# This is a valid program but it can be made simpler if we are able to write condition or criteria that breaks the loop on top
#this way we don't have to check for that condition with every other statement.

while (True):
    generated_number=random.randint(1,100)
    user_input=int(input('''\nTo Play Enter Number Between 1-100:
    Press '0' to Quit: '''))
    
    if user_input>0 and user_input==generated_number:
        print('Congratulations you have won')
        print('Generated Number:',generated_number)
    elif user_input>0 and user_input<generated_number:
        print('Your guess is on lower side')
        print('Generated Number:',generated_number)
    elif user_input>0 and user_input>generated_number:
        print('Your guess is on higher side')
        print('Generated Number:',generated_number)
    elif user_input == 0:
        print('Quiting Game')
        break;
```
