# Python Fundamentals:

### 1. In Python everything is an OBJECT (an object has: its type, its value, and reference count).
### 2. We assign variables (i.e names) to an object unlike in C, C++, Java where we have to declare a variable with its type before assigning a value.
### 2. In Python there are NO variables just names that are assigned to an object. 
### 3. An object is stored in memory with names and refernces. A name is just a label for an object so, an object can have many names. A reference is a name (pointer) that references to the original object.


Consider following code for better understanding:

```python
a=50
b=a
c=50
```

```python
print(id(a))
print(id(b))
print(id(c))
```

* As we can see ther is only object that has been allocated a memory and has an id to refer. The id() function returns a unique id for the specified object. All objects in Python has its own unique id. The id is the object's memory address, and will be different for each time you run the program.

* This shows that objects (values) are not just copied into another varibles in Python, we have to explicitly tell Python to make a new copy (new object and assign the variable/name to this object).

```python
print (a is b)
print (b is c)
print (c is a)
```

 * The result above shows that all the variables (names) point to the same object at the above id (memory location).

# Memory Allocation:
 ### Memory Management (allocation & deallocation) in Python is automatic. Python detects type of an object when it is declared and a varible is assigned to it, memory allocation is done during runtime.
 ### When an object is no longer in use (i.e it doesn't have names or references (its reference count is zero), the memory space is freed-up by Garbage Collector. A Garbage Collector keeps track of all objects in memory. 
 ### Garbage Collector frees up memory in two ways:
> * Reference Count of Object: If it reaches zero, object is removed\
> * Generational Garbage Collection: There are two concepts in this - Generation (3 level), Threshold.\
A new object starts its life in first generation of GC if Py executes a GC process on a genration and an object survives they are moved to second generation, and so on. In Threshold we set a threshold for an object to survive a GC process.\
